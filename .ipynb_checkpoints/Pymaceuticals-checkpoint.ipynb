## Observations and Insights 

Based on the data below, the following can be discerned:

1) Over the 45 days, Capomulin and Ramicane had significantly lower mean tumor volume numbers compared to the rest of the treatments administered, meaning that these two drugs were more effective at eliminating the tumors than the other treatments. Also, Capomulin and Ramicane also had lower variance, standard deviation, REM values which shows that the results across different mice were able to be replicated to a relatively similar degree.

2) The bar charts show that the ratio mice were relatively equal among the tests performed, and the pie charts show the ratio of male to female mice were about equal.

3) Our box plots show that among the drugs Capomulin, Ramicane, Infubinol, and Ceftamin, the majority of the final tumor volume data falls within the upper and lower bounds. The only exception is one outlier for Infubinol which falls slightly under the lower bound of the plot.

4) To showcase tumor volume vs. timepoint for Capomulin, mouse b128 was chosen for line plot. Based on the chart, the tumor volume significantly decreases as the timepoint increases, but interestingly shoots up slightly towards the end.

5) The final scatterpoint shows a positive correlation between average tumor volume and mouse weight. 

## Dependencies and Setup

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import sem
import numpy as np
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
total_data_df = mouse_metadata.merge(study_results, how="outer", on="Mouse ID")

# Display the data table for preview
total_data_df.head()
```

```python
# Checking the number of mice.
mice_count = total_data_df["Mouse ID"].nunique()
print(f"Total number of mice: {mice_count}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
# Optional: Get all the data for the duplicate mouse ID. 

duped_data = total_data_df.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)
total_data_df.loc[duped_data]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_total_data_df = total_data_df.loc[total_data_df['Mouse ID'] != 'g989']
```

```python
clean_total_data_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_number = clean_total_data_df['Mouse ID'].nunique()
print(f'Total number of mice after dropping the duplicate mouse data: {clean_number}')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

drug_df = clean_total_data_df[['Tumor Volume (mm3)', 'Drug Regimen']]
drug_grouped_df = drug_df.groupby('Drug Regimen')

drug_mean = drug_grouped_df['Tumor Volume (mm3)'].mean()
drug_median = drug_grouped_df['Tumor Volume (mm3)'].median()
drug_variance = drug_grouped_df['Tumor Volume (mm3)'].var()
drug_std = drug_grouped_df['Tumor Volume (mm3)'].std()
drug_sem = drug_grouped_df['Tumor Volume (mm3)'].sem()

summary_drug_df = pd.DataFrame({'Tumor Volume Mean': drug_mean,
                               'Tumor Volume Median': drug_median,
                               'Tumor Volume Variance': drug_variance,
                               'Tumor Volume Standard Dev': drug_std,
                              'Tumor Volume SEM': drug_sem})
summary_drug_df.head()
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.
mouse_data = clean_total_data_df[['Mouse ID', 'Drug Regimen']]
unique_mouse_data = mouse_data.drop_duplicates(subset=['Mouse ID'])
unique_mouse_data = unique_mouse_data.groupby("Drug Regimen")

pandasbar = unique_mouse_data.count().plot.bar(color="c", title="Drug Regimen v Number of Unique Mice Tested")
pandasbar.set_xlabel("Drug Regimen")
pandasbar.set_ylabel("Number of Unique Mice Tested")
pandasbar.legend(["Number of Unique Mice"]);
plt.tight_layout()

plt.savefig('pandasbarchart.png')
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.

unique_mouse_df = pd.DataFrame(unique_mouse_data.count())

x_axis = np.arange(len(unique_mouse_df))
tick_locations = [x for x in x_axis]

plt.figure()
plt.bar(x_axis, unique_mouse_df["Mouse ID"], color='c', align="center", width = 0.50)
plt.xticks(tick_locations, list(unique_mouse_df.index), rotation="vertical")

plt.title("Drug Regimen v Number of Unique Mice Tested")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Unique Mice Tested")
plt.legend(["Number of Unique Mice"]);
plt.tight_layout()

plt.show()
plt.savefig('pyplotbarchart.png')
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas

gender_data = clean_total_data_df[['Mouse ID', 'Sex']]
gender_mouse_data = gender_data.drop_duplicates(subset=['Mouse ID'])
gender_mouse_data = gender_mouse_data.groupby('Sex')
gender_mouse_df = pd.DataFrame(gender_mouse_data['Sex'].count())

pandaspie = gender_mouse_df.plot(kind='pie', y='Sex', 
                        title="Distribution of Female and Male Mice",startangle=120,
                        autopct='%1.1f%%', colors =["pink","lightblue"],legend =False)
pandaspie

plt.savefig('pandaspie.png')
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = clean_total_data_df.loc[(clean_total_data_df["Timepoint"])]
last_timepoint_df = clean_total_data_df.groupby(['Mouse ID'])['Timepoint'].max()
last_timepoint_df = last_timepoint_df.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merged_last_timepoint_df = clean_total_data_df.merge(last_timepoint_df, how="right", on=["Mouse ID", "Timepoint"])

merged_last_timepoint_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Determine outliers using upper and lower bounds
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

tumor_data = []

for drug in drug_list:
    tumor_vol_df = merged_last_timepoint_df.loc[merged_last_timepoint_df["Drug Regimen"]==drug,"Tumor Volume (mm3)"]
    
    quartiles = tumor_vol_df.quantile([0.25,0.5,0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    lower_b = lowerq - (1.5*iqr)
    upper_b = upperq + (1.5*iqr)
    
    print(f'{drug}: Tumor volume has an upper bound of {upper_b} and a lower bound of {lower_b}')
    
    outlier = tumor_vol_df.loc[(tumor_vol_df>upper_b)|(tumor_vol_df<lower_b)]
    print(f'{drug}: Outliers: {outlier}')
    
    fig1, ax1 = plt.subplots()
    ax1.set_title(f' Final Tumor Volume for Mice Treated With {drug}')
    ax1.set_ylabel('Tumor Volume (mm3)')
    ax1.boxplot(tumor_vol_df)
    plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

one_mouse = clean_total_data_df.loc[clean_total_data_df["Mouse ID"] == "b128", ["Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]

one_mouse_timepoint = one_mouse["Timepoint"]
one_mouse_tumorvol = one_mouse["Tumor Volume (mm3)"]

plt.figure()
plt.plot(one_mouse_timepoint, one_mouse_tumorvol) 

plt.title("Time point v Tumor Volume for mouse B128")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")

plt.savefig('onemouseline.png')
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

one_mouse_df = clean_total_data_df.loc[clean_total_data_df["Drug Regimen"] == "Capomulin"]
one_mouse_vol_df = one_mouse_df.groupby("Mouse ID").mean()["Tumor Volume (mm3)"]
one_mouse_weight_df = one_mouse_df.groupby("Mouse ID").mean()["Weight (g)"]

one_mouse_weight_df

```

```python
plt.figure()
plt.scatter(one_mouse_vol_df, one_mouse_weight_df)
plt.xlabel("Tumor Volume (mm3)")
plt.ylabel("Weight (g)")
plt.title("Avg Tumor Volume v Mouse Weight")
plt.show()

plt.savefig('avgtumorvmouseweightscatter.png')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

print(f"The correlation coefficient between mouse weight and average tumor volume is {round(st.pearsonr(one_mouse_vol_df, one_mouse_weight_df)[0],2)}")
```

```python
t_slope, t_int, t_r, t_p, t_std_err = stats.linregress(one_mouse_vol_df, one_mouse_weight_df)
t_fit = t_slope * one_mouse_vol_df + t_int

plt.figure()
plt.scatter(one_mouse_vol_df, one_mouse_weight_df)
plt.plot(one_mouse_vol_df, t_fit,"--")
plt.xlabel("Tumor Volume (mm3)")
plt.ylabel("Weight (g)")
plt.title("Avg Tumor Volume v Mouse Weight")
plt.show()

plt.savefig('avgtumorvolvmouseweightregression.png')
```
